```python
import numpy as np
import tensorflow as tf
import vgg16
import utils
import cv2
```

```python
np_load_old = np.load

np.load = lambda *a, **k: np_load_old(*a, allow_pickle=True, **k)
```

```python
img = utils.load_image("./test_data/tiger.jpeg")

batch = img.reshape((1, 224, 224, 3))

# with tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(gpu_options=(
# tf.compat.v1.GPUOptions(per_process_gpu_memory_fraction=0.7)))) as sess:
with tf.device('/cpu:0'):
    with tf.compat.v1.Session() as sess:
        images = tf.compat.v1.placeholder("float", [1, 224, 224, 3])
        feed_dict = {images: batch}

        vgg = vgg16.Vgg16()
        with tf.compat.v1.name_scope("content_vgg"):
            vgg.build(images)

        prob = sess.run(vgg.prob, feed_dict=feed_dict)
        print("Top 5 object gan giong voi hinh tiger.jpeg:")
        utils.print_prob(prob[0], './synset.txt')
```

```python
img1 = utils.load_image("./test_data/dog-and-cat.jpg")
img2 = utils.load_image("./test_data/elephant-and-dog.jpg")

batch1 = img1.reshape((1, 224, 224, 3))
batch2 = img2.reshape((1, 224, 224, 3))

batch = np.concatenate((batch1, batch2), 0)

# with tf.Session(config=tf.ConfigProto(gpu_options=(tf.GPUOptions(per_process_gpu_memory_fraction=0.7)))) as sess:
with tf.compat.v1.device('/cpu:0'):
    with tf.compat.v1.Session() as sess:
        images = tf.compat.v1.placeholder("float", [2, 224, 224, 3])
        feed_dict = {images: batch}

        vgg = vgg16.Vgg16()
        with tf.compat.v1.name_scope("content_vgg"):
            vgg.build(images)

        prob = sess.run(vgg.prob, feed_dict=feed_dict)
        print("Top 5 object gan giong voi hinh dog-and-cat:")
        utils.print_prob(prob[0], './synset.txt')
        print("Top 5 object gan giong voi hinh elephant-and-dog:")
        utils.print_prob(prob[1], './synset.txt')
```

```python
img1 = utils.load_image("./test_data/tiger.jpeg")
img2 = utils.load_image("./test_data/tiger_crop_unimportant.jpg")
img3 = utils.load_image("./test_data/tiger_crop_important.jpg")

batch1 = img1.reshape((1, 224, 224, 3))
batch2 = img2.reshape((1, 224, 224, 3))
batch3 = img3.reshape((1, 224, 224, 3))

batch = np.concatenate((batch1, batch2, batch3), 0)

# with tf.Session(config=tf.ConfigProto(gpu_options=(tf.GPUOptions(per_process_gpu_memory_fraction=0.7)))) as sess:
with tf.compat.v1.device('/cpu:0'):
    with tf.compat.v1.Session() as sess:
        images = tf.compat.v1.placeholder("float", [3, 224, 224, 3])
        feed_dict = {images: batch}

        vgg = vgg16.Vgg16()
        with tf.compat.v1.name_scope("content_vgg"):
            vgg.build(images)

        prob = sess.run(vgg.prob, feed_dict=feed_dict)
        print("Top 5 object gan giong voi hinh tiger:")
        utils.print_prob(prob[0], './synset.txt')
        print("Top 5 object gan giong voi hinh tiger_crop_unimportant:")
        utils.print_prob(prob[1], './synset.txt')
        print("Top 5 object gan giong voi hinh tiger_crop_important:")
        utils.print_prob(prob[2], './synset.txt')
```

```python
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def imshow(*imgs, dpi=100):
    fig=plt.figure(figsize=(8, 8), dpi=dpi)
    columns = len(imgs)
    rows = 1
    for i in range(1, columns*rows +1):
        fig.add_subplot(rows, columns, i)
        if len(imgs[i-1].shape) == 2:
            plt.imshow(imgs[i-1], cmap='gray')
        else:
            plt.imshow(imgs[i-1])
```

```python
img = utils.load_image("./test_data/tiger.jpeg")

batch = img.reshape((1, 224, 224, 3))

# with tf.Session(config=tf.ConfigProto(gpu_options=(tf.GPUOptions(per_process_gpu_memory_fraction=0.7)))) as sess:
with tf.compat.v1.device('/cpu:0'):
    with tf.compat.v1.Session() as sess:
        images = tf.compat.v1.placeholder("float", [1, 224, 224, 3])
        feed_dict = {images: batch}

        vgg = vgg16.Vgg16()
        with tf.compat.v1.name_scope("content_vgg"):
            vgg.build(images)

        prob, conv1_1, conv2_1, conv3_1, conv4_1, conv5_1 = \
        sess.run([vgg.prob, vgg.conv1_1, vgg.conv2_1, 
                  vgg.conv3_1, vgg.conv4_1, vgg.conv5_1], feed_dict=feed_dict)
        # prob, conv2_2 = sess.run([vgg.prob, vgg.conv2_1], feed_dict=feed_dict)

        conv1_1 = conv1_1[0, :, :, 0] # featuremap 1
        conv2_1 = conv2_1[0, :, :, 48] # featuremap 48
        conv3_1 = conv3_1[0, :, :, 40]# featuremap 40
        conv4_1 = conv4_1[0, :, :, 20] # freaturemap 20
        conv5_1 = conv5_1[0, :, :, 30] # featuremap 30
      
        imshow(conv1_1, conv2_1, conv3_1, conv4_1, conv5_1)
        
        # cv2.waitKey(0)
        # cv2.destroyAllWindows()

        utils.print_prob(prob[0], './synset.txt')
```
